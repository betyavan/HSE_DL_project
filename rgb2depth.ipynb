```python
def image_to_torch(image, device=None):
    """Convert Numpy image to Torch tensor."""
    if isinstance(image, torch.Tensor):
        return image
    if image.shape[-1] == 4:
        image = image[..., :3]
    result = torch.from_numpy(np.ascontiguousarray(image))
    if device is not None:
        result = result.to(device)
    #if len(result.shape) == 3:
    result = result.permute(2, 0, 1)
    return result

  
def image_to_numpy(image):
    """Convert Torch tensor to Numpy image."""
    if isinstance(image, np.ndarray):
        return image
    if image.dtype == torch.float32:
        image = image - image.min()
        image = 255 * image / max(1e-6, image.max())
    #if len(image.shape) == 3:
    return image.to(torch.uint8).permute(1, 2, 0).cpu().numpy()
    #return image.to(torch.uint8).cpu().numpy()
```

```python
#import torchvision.transforms.RandomHorizontalFlip
import torchvision.transforms as tf

train_transforms =  tf.Compose([
            tf.ToPILImage(), 
            tf.Resize(size=(256, 256)),
            tf.ToTensor()
])

val_transforms =  tf.Compose([
            tf.ToPILImage(), 
            tf.Resize(size=(256, 256)),
            tf.ToTensor(),  
            #tf.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
            ])
    
class FlipAugmenter(object):
    def __init__(self, p=0.5):
        self.p = p
    
    def __call__(self, image, mask):
        if np.random.rand() > self.p:
            image = np.flip(image, axis=1)
            mask = np.flip(mask, axis=1)
        return image, mask

class SYNTHIADataset(Dataset):
    def __init__(self, imgs_path, transforms):
        super(SYNTHIADataset, self).__init__()
        self.imgs_path = imgs_path
        self.transforms = transforms
        self.augmenter = FlipAugmenter()
        
    def __len__(self):
        return len(self.imgs_path)
    
    def __getitem__(self, idx):
        rgb = cv2.imread(self.imgs_path[idx][0], cv2.IMREAD_COLOR)
        rgb = cv2.cvtColor(rgb, cv2.COLOR_BGR2RGB)
        depth = cv2.imread(self.imgs_path[idx][1], cv2.IMREAD_GRAYSCALE)
            
        rgb, depth = self.augmenter(rgb, depth)
        rgb = self.transforms(rgb)
        depth = self.transforms(depth)
        
        return rgb, depth
        
```

```python
np.random.shuffle(imgs_path)
n = len(imgs_path)
train_size = 0.8

train_path = imgs_path[:int(n * train_size)]
val_path = imgs_path[int(n * train_size):]

train_dataset = SYNTHIADataset(train_path, train_transforms)
val_dataset = SYNTHIADataset(val_path, val_transforms)
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)
```

### Models

```python
class Discriminator(nn.Module):
    def __init__(self, input_shape):
        super(Discriminator, self).__init__()

        channels, height, width = input_shape

        # Calculate output shape of image discriminator (PatchGAN)
        self.output_shape = (1, height // 2 ** 4, width // 2 ** 4)

        def discriminator_block(in_filters, out_filters, normalize=True):
            """Returns downsampling layers of each discriminator block"""
            layers = [nn.Conv2d(in_filters, out_filters, 4, stride=2, padding=1)]
            if normalize:
                layers.append(nn.InstanceNorm2d(out_filters))
            layers.append(nn.LeakyReLU(0.2, inplace=True))
            return layers

        self.model = nn.Sequential(
            *discriminator_block(channels, 64, normalize=False),
            *discriminator_block(64, 128),
            *discriminator_block(128, 256),
            *discriminator_block(256, 512),
            nn.ZeroPad2d((1, 0, 1, 0)),
            nn.Conv2d(512, 1, 4, padding=1)
        )
        
        self.fc = nn.Sequential(
            nn.Linear(256, 50, bias=False),
            nn.ReLU(),
            nn.Linear(50, 1)
        )

    def forward(self, img):
        x = self.model(img)
        x = x.view(x.shape[0], -1)
        x = self.fc(x)
        #print(x.shape)
        return x
```

```python
device = 'cuda'

generator = torch.hub.load('mateuszbuda/brain-segmentation-pytorch', 'unet',
    in_channels=3, out_channels=1, init_features=32, pretrained=False)

discriminator = Discriminator((1, 256, 256))

generator.to(device)
discriminator.to(device)

beta_1 = 0.5
beta_2 = 0.999
lr = 0.0002
optim_G = optim.Adam(params=generator.parameters(), lr=lr, betas=(beta_1, beta_2))
optim_D = optim.Adam(params=discriminator.parameters(), lr=lr, betas=(beta_1, beta_2))
```

### Training

```python
def gradient_of_discriminator_score(discriminator, real, fake, epsilon):
    interpolated_images = real * epsilon + fake * (1 - epsilon)
    mixed_scores = discriminator(interpolated_images).squeeze(-1)

    gradient = torch.autograd.grad(
        inputs=interpolated_images,
        outputs=mixed_scores,
        grad_outputs=torch.ones_like(mixed_scores),
        create_graph=True,
        retain_graph=True,
    )[0]
    
    return gradient

def gradient_penalty_l2_norm(gradient):
    gradient = gradient.view(len(gradient), -1)
    gradient_norm = gradient.norm(2, dim=1)
    penalty = torch.mean((gradient_norm - 1) ** 2)
    return penalty
```

```python
from tqdm.notebook import tqdm
from IPython.display import clear_output
```

```python
fixed_rgbs, fixed_depths = next(iter(train_dataloader))
num_watch = 3
fixed_rgbs = fixed_rgbs[:num_watch].to(device)
fixed_depths = fixed_depths[:num_watch].to(device)

num_epochs = 1
rec_loss = nn.L1Loss()
coeff = 0.99

Lambda = 10
n_critic = 2

ERRD = np.zeros(num_epochs)
ERRG = np.zeros(num_epochs)
N = len(train_dataloader)

for epoch in range(num_epochs):
    for iteration, (rgbs, depths) in enumerate(tqdm(train_dataloader)):
        discriminator.zero_grad()
        # real
        output_real = discriminator(depths.to(device)).squeeze(-1)
        
        # fake 
        fake = generator(rgbs.to(device))
        output_fake = discriminator(fake.detach()).squeeze(-1)
        
        # Gradient Penalty
        
        epsilon = torch.rand(len(rgbs), 1, 1, 1, requires_grad=True).to(device)
        gradient = gradient_of_discriminator_score(discriminator, depths.to(device),
                                                   fake.detach(), epsilon)
        gp = gradient_penalty_l2_norm(gradient)
        errD = torch.mean(output_fake - output_real + Lambda * gp)
        ERRD[epoch] += errD.item()
        errD.backward()
        optim_D.step()
        
        ####### 
        # Generator 
        #######
        if (iteration + 1) % n_critic == 0:
            generator.zero_grad()
            output = discriminator(fake).squeeze(-1)
            errG_adv = -output.mean()
            errG_rec = rec_loss(fake, depths.to(device))
            errG = coeff * errG_rec + (1 - coeff) * errG_adv
            ERRG[epoch] += errG.item()
            errG.backward()
            optim_G.step()
          
        if (iteration+1) % 3 == 0:
            fake = generator(fixed_rgbs)
            #plt.figure(figsize=(6, 7))
            for i in range(num_watch):
                plt.subplot(num_watch, 3, num_watch*i + 1)
                plt.imshow(image_to_numpy(fixed_rgbs[i]))
                plt.axis('off')
                
                plt.subplot(num_watch, 3, num_watch*i+2)
                plt.imshow(image_to_numpy(fake[i]))
                plt.axis('off')
                
                plt.subplot(num_watch, 3, num_watch*i+3)
                plt.imshow(image_to_numpy(fixed_depths[i]))
                plt.axis('off')
            clear_output(True)
            print('Epoch:{} Iter: {} errD: {:.2f} errG: {:.2f}'.format(epoch+1,iteration+1,
                                                                       errD.item(), errG.item()))
            plt.show()
            
ERRD = np.array(ERRD) / N
ERRG = np.array(ERRG) / N * n_critic
```
